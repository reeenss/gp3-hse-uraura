```python
import numpy as np
import pandas as pd

# Приводим все эмоции в треке и эмоции из df_fixed к нижнему регистру и убираем пробелы
tracks['emotions_clean'] = tracks['emotions'].apply(
    lambda x: [e.strip().lower() for e in x]
)
df_fixed['Emotion_clean'] = df_fixed['Emotion'].apply(
    lambda x: x.strip().lower()
)

# Функция для поиска валентности и активации для каждой эмоции
# и расчёта трековых метрик
def calculate_track_emotions(track_emotions):
    track_valence = []
    track_arousal = []

    # Для каждой эмоции из трека находим соответствующие значения валентности и активации
    for emotion in track_emotions:
        emotion_data = df_fixed[df_fixed['Emotion_clean'] == emotion]
        if not emotion_data.empty:
            track_valence.append(float(emotion_data['Valence'].values[0]))
            track_arousal.append(float(emotion_data['Arousal'].values[0]))

    # Если ничего не нашли — возвращаем нейтральные значения
    if not track_valence:
        return 50.0, 50.0, 0.0, 0.0

    V = np.array(track_valence)
    A = np.array(track_arousal)

    # 1) Взвешенная по активации валентность (важнее эмоции с высокой энергией)
    if A.sum() > 0:
        avg_valence = (V * A).sum() / A.sum()
    else:
        avg_valence = V.mean()

    # 2) Средняя активация
    avg_arousal = A.mean()

    # 3) Индекс мании трека (высокая активация + экстремальность валентности)
    v_norm = V / 100.0
    a_norm = A / 100.0
    mania_emotions = a_norm * (1.0 + np.abs(v_norm - 0.5))
    mania_index = mania_emotions.mean() * 100.0  # 0–100

    # 4) Индекс депрессии трека (низкая валентность + не слишком высокая активация)
    depression_emotions = (1.0 - v_norm) * (1.0 - 0.5 * a_norm)
    depression_index = depression_emotions.mean() * 100.0  # 0–100

    return avg_valence, avg_arousal, mania_index, depression_index

# Применяем функцию для каждого трека
tracks[['avg_valence', 'avg_arousal', 'mania_index', 'depression_index']] = (
    tracks['emotions_clean']
    .apply(calculate_track_emotions)
    .apply(pd.Series)
)

# Таблица для анализа эмоций по трекам
em_tr = tracks[['track_title', 'avg_valence', 'avg_arousal', 'mania_index', 'depression_index']]
em_tr.head()
```

```python
import numpy as np
import pandas as pd

def calc_track_emo_coefficient(row):
    v = row['avg_valence_x']
    a = row['avg_arousal_x']
    mania = row['mania_index_x']
    depression = row['depression_index_x']

    if pd.isna(v) or pd.isna(a):
        return np.nan

    # 1) нормировка валентности и активации в [-1; 1]
    Vn = (v - 50) / 50.0
    An = (a - 50) / 50.0

    # 2) базовая эмоциональная энергия
    energy = Vn * 0.6 + An * 0.8

    # 3) нелинейное усиление
    nonlinear = energy * abs(An)

    EMO_score = 100 * nonlinear
    EMO_score = max(-100, min(100, EMO_score))

    # 4) влияние мании и депрессии
    mania_norm = (mania - 50) / 50.0 if not pd.isna(mania) else 0.0
    depr_norm = (depression - 50) / 50.0 if not pd.isna(depression) else 0.0

    bipolar_factor = 1 + 0.4 * mania_norm - 0.4 * depr_norm

    final_score = EMO_score * bipolar_factor
    final_score = max(-150, min(150, final_score))

    return final_score   # ← важно, а то функция возвращает None

tracks_full['track_emo_score'] = tracks_full.apply(
    calc_track_emo_coefficient, axis=1
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# берём треки и сразу переименовываем нужные колонки *_x -> без суффикса
df = tracks_full.copy()

df = df.rename(columns={
    'avg_valence_x': 'avg_valence',
    'avg_arousal_x': 'avg_arousal',
    'mania_index_x': 'mania_index',
    'depression_index_x': 'depression_index'
})

def calc_track_emo_v2(row):
    v = row['avg_valence']
    a = row['avg_arousal']
    mania = row['mania_index']
    depression = row['depression_index']

    # если чего-то нет — NaN
    if pd.isna(v) or pd.isna(a) or pd.isna(mania) or pd.isna(depression):
        return np.nan

    # 1. нормировка в [-1; 1]
    Vn = (v - 50.0) / 50.0
    An = (a - 50.0) / 50.0
    Mn = (mania - depression) / 50.0      # баланс мания–депрессия
    Mn = np.clip(Mn, -1, 1)

    # 2. базовое "биполярное" ядро
    core = 0.4 * An - 0.3 * Vn + 0.6 * Mn

    # 3. усиление по общей интенсивности индексов
    intensity = (mania + depression) / 200.0   # 0..1
    factor = 0.5 + 0.5 * intensity             # 0.5..1

    score = 100.0 * core * factor

    # клип в [-100; 100]
    score = np.clip(score, -100, 100)
    return score

# пересчитываем
df['track_emo_score_v2'] = df.apply(calc_track_emo_v2, axis=1)

# гистограмма новой метрики
scores = df['track_emo_score_v2'].dropna()

plt.figure(figsize=(10, 6))
plt.hist(scores, bins=60, edgecolor='black', alpha=0.7)

mean_score = scores.mean()
plt.axvline(mean_score, linestyle='--', linewidth=2,
            label=f'Среднее = {mean_score:.1f}')

plt.title('Распределение нового коэффициента трека (track_emo_score_v2)')
plt.xlabel('track_emo_score_v2')
plt.ylabel('Количество треков')
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```
