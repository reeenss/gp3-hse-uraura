```python
# Приводим все эмоции в треке и эмоции из df_fixed к нижнему регистру и убираем пробелы
tracks['emotions_clean'] = tracks['emotions'].apply(lambda x: [e.strip().lower() for e in x])  # очищаем эмоции для трека
df_fixed['Emotion_clean'] = df_fixed['Emotion'].apply(lambda x: x.strip().lower())  # очищаем эмоции в df_fixed

# Функция для поиска валентности и активации для каждой эмоции
def calculate_track_emotions(track_emotions):
    track_valence = []
    track_arousal = []

    # Для каждой эмоции из трека находим соответствующие значения валентности и активации
    for emotion in track_emotions:
        # Ищем совпадение по очищенным эмоциям
        emotion_data = df_fixed[df_fixed['Emotion_clean'] == emotion]
        if not emotion_data.empty:
            track_valence.append(emotion_data['Valence'].values[0])
            track_arousal.append(emotion_data['Arousal'].values[0])

    # Рассчитываем среднее значение для валентности и активации
    avg_valence = sum(track_valence) / len(track_valence) if track_valence else 50
    avg_arousal = sum(track_arousal) / len(track_arousal) if track_arousal else 50

    return avg_valence, avg_arousal

# Применяем функцию для каждого трека и получаем значения валентности и активации
tracks[['avg_valence', 'avg_arousal']] = tracks['emotions_clean'].apply(calculate_track_emotions).apply(pd.Series)

# Выводим результат
em_tr = tracks[['track_title', 'avg_valence', 'avg_arousal']]
em_tr.head()
```
