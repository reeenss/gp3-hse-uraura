Ключевая метрика: конверсия регистрации

В качестве основной метрики эксперимента выбрана конверсия в регистрацию. Это метрика, которая непосредственно соответствует целям тестируемых изменений, так как направлена на увеличение числа зарегистрированных пользователей и отражает первый ключевой шаг продуктовой воронки.

# Гипотезы

**Нулевая гипотеза:** Конверсия в контроле равна конверсии в тесте

**Альтернативная:** Конверсия в контроле не равна конверсии в тесте
